```python
import torch
from win32print import PRINTER_ACCESS_ADMINISTER
```

```python
# detach()方法演示
# x: 带梯度计算的张量，requires_grad=True
# y: 通过detach()分离得到的新张量，不参与梯度计算
x=torch.tensor(2.0,requires_grad=True)
y=x.detach()
print('原始张量 x:', x)
print('分离张量 y:', y)
```

```python
# 检查梯度计算属性
# requires_grad: True表示参与自动微分，False表示不参与
# detach()会创建新的张量，且requires_grad=False
print('x.requires_grad:', x.requires_grad)
print('y.requires_grad:', y.requires_grad)
```

```python
print(id(x))
print(id(y))
```

```python
print(x.untyped_storage().data_ptr())
print(y.untyped_storage().data_ptr())
```

```python
# 分别对x和y进行后续计算
z1=x**2
z2=y**2
print(z1)
print(z2)
```

```python
z1.sum().backward()
```

```python
x=torch.ones(2,2,requires_grad=True)
y=x*x
print(x)
print(y)
```

```python
# 分离y
u = y.detach()
print(u)
```

```python
# 让u参与新张量计算
z=u*x
```

```python
# 反向传播，计算所有梯度
z.sum().backward()
```

```python
print(x.grad==u)
```

```python
# 卷积层的定义和应用测试
import torch
import matplotlib.pyplot as plt
```

```python
# 1. 读取图片
img=plt.imread("../data/duck.jpg")
# 2. 将图片数据调整为卷积层输入特征图对应的形状
input=torch.tensor(img).permute(2,0,1).float()
print(input.shape)
```

```python
# 3.定义卷积层
conv=torch.nn.Conv2d(in_channels=3,out_channels=3,kernel_size=9,stride=3,padding=0)
```

```python
# 4. 前向传播，将卷积层应用到输入特征图上
output=conv(input)
print(output.shape)
```

```python
# 5. 定义池化层
pool=torch.nn.MaxPool2d(kernel_size=6,stride=6,padding=1)
```

```python
# 6. 前向传播，池化操作
output2=pool(output)
print(output2.shape)
```

```python
# 7. 将输出特征图转换为图片数据
output=torch.clamp(output.int(),min=0,max=255)
output=output.permute(1,2,0).detach().numpy()
output2=torch.clamp(output2.int(),min=0,max=255)
output2=output2.permute(1,2,0).detach().numpy()
```

```python
# 显示图片，进行对比
fig,ax=plt.subplots(1,3,figsize=(10,5))
ax[0].imshow(img)
ax[1].imshow(output)
ax[2].imshow(output2)
plt.show()
```

```python
import torch
import torch.nn as nn
```

```python
input = torch.randn(3,5)
target = torch.randn(3,5)
print(input)
print(target)
```

```python
# MSE:L2 loss
mse_loss= nn.MSELoss()
print(mse_loss(input, target))
```

```python
# MAE: L1 Loss
mae_loss = nn.L1Loss()
print(mae_loss(input, target))
```

```python
# Smooth L1
smooth_l1_loss = nn.SmoothL1Loss()
print(smooth_l1_loss(input, target))
```

```python
import torch
import torch.nn as nn
```

```python
# 定义全连接层
linear= nn.Linear(5,2)
```

```python
# 1. 常数初始化
nn.init.zeros_(linear.weight)
print(linear.weight)
```

```python
nn.init.ones_(linear.weight)
print(linear.weight)
```

```python
nn.init.constant_(linear.weight, 10)
print(linear.weight)
```

```python
# 2. 秩初始化
nn.init.eye_(linear.weight)
print(linear.weight)
```

```python
# 3.正态分布初始化
nn.init.normal_(linear.weight, mean=0, std=0.01)
print(linear.weight)
```

```python
# 4. 均匀分布初始化
nn.init.uniform_(linear.weight, a=-3, b=3)
print(linear.weight)
```

```python
# 5.Xavier初始化
nn.init.xavier_uniform_(linear.weight)
print(linear.weight)
```

```python
nn.init.xavier_normal_(linear.weight)
print(linear.weight)
```

```python
# He初始化
nn.init.kaiming_uniform_(linear.weight,a=0)
print(linear.weight)
```

```python
nn.init.kaiming_normal_(linear.weight,a=0.01)
print(linear.weight)
```
